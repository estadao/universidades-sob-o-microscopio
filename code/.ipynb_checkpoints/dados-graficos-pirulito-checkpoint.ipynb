Esse script gera o esqueleto em SVG dos gráficos "pirulito" da matéria usando `matplotlib` e `pandas.

#### Importar pacotes

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

#### Ler dados

```python
with pd.ExcelFile("../data/CWTS Leiden Ranking 2019.xlsx") as xls:
    df = pd.read_excel(xls, 'Results')
```

#### Filtrar dataframe

```python
df = df [ (df.Period == "2014–2017") &
          (df.Country == "Brazil") &
          (df.Frac_counting == 1) ]
```

#### Funções auxiliares

```python
def rename_field(field):
    '''
    Altera o padrão de nomenclatura 
    dos campos de conhecimento.
    '''

    corresp = {
        "Biomedical and health sciences" : "Ciências biomédicas e da saúde",
        "Life and earth sciences" : "Ciências da vida e da terra",
        "Physical sciences and engineering" : "Ciências físicas e engenharia",
        "Social sciences and humanities" : "Ciências sociais e humanidades",
        "Mathematics and computer science" : "Matemática e ciência da computação"
    }

    return corresp[field]
```

```python
def rename_univ(univ):
    '''
    Altera o padrão de nomenclatura
    das universidades brasileiras.
    '''

    corresp = {
         'Federal University of Paraíba': 'UFPB – Univ. Fed. da Paraíba',
         'Federal University of Pernambuco': 'UFPE – Univ. Fed. de Pernambuco',
         'Federal University of Rio Grande do Norte': 'UFRN – Univ. Fed. do Rio Grande do Norte',
         'Federal University of Rio Grande do Sul': 'UFRGS – Univ. Fed. do Rio Grande do Sul',
         'Federal University of Santa Maria': 'UFSM – Univ. Fed. de Santa Maria',
         'Federal University of São Carlos': 'Ufscar – Univ. Fed. de São Carlos',
         'Federal University of Uberlândia': 'UFU – Univ. Fed. de Uberlândia',
         'Rio de Janeiro State University': 'UERJ – Univ. Est. do Rio de Janeiro',
         'State University of Maringá': 'UEM – Univ. Est. de Maringá',
         'Universidade Estadual Paulista': 'Unesp – Univ. Est. Paulista',
         'Universidade Federal Fluminense': 'UFF – Univ. Fed. Fluminense ',
         'Universidade Federal da Bahia': 'UFBA – Univ. Fed. da Bahia',
         'Universidade Federal de Goiás': 'UFGO – Univ. Fed. de Goiás',
         'Universidade Federal de Minas Gerais': 'UFMG – Univ. Fed. de Minas Gerais',
         'Universidade Federal de Santa Catarina': 'UFSC – Univ. Fed. de Santa Catarina',
         'Universidade Federal de São Paulo': 'Unifesp – Univ. Fed. de São Paulo',
         'Universidade Federal de Viçosa': 'UFV – Univ. Fed. de Viçosa',
         'Universidade Federal do Ceará': 'UFC – Univ. Fed. do Ceará',
         'Universidade Federal do Paraná': 'UFPR - Univ. Fed. do Paraná',
         'Universidade Federal do Rio de Janeiro': 'UFRJ – Univ. Fed. do Rio de Janeiro',
         'Universidade de Brasília': 'UnB – Univ. de Brasília',
         'University of Campinas': 'Unicamp – Univ. Est. de Campinas',
         'University of São Paulo': 'USP – Univ. de São Paulo'
    }

    return corresp[univ]
```

```python
def rename_col(col):
    '''
    Altera o padrão de nomenclatura
    das universidades brasileiras.
    '''

    corresp = {
        'PP_OA' : 'Percentual de artigos de acesso aberto',
        'PA_F_MF' : 'Percentual de pesquisadoras mulheres',
        'PP_int_collab' : 'Percentual de artigos feitos em colaboração com universidadades estrangeiras'
    }

    return corresp[col]
```

```python
for index, row in df.iterrows():
    df.loc[index, "University"] = rename_univ(row.University)
```

#### Gráficos 1 – Áreas do conhecimento

```python
# Precisamos de um gráfico para cada campo que queremos destacar
fields = [ field for field in df.Field.unique() if field != "All sciences"]
for field in fields:
    
    # Filtra o dataframe
    temp = df [ df.Field == field]
    temp = temp.sort_values(by = "PP_top10", ascending = False)
    temp = temp.head(5).sort_values(by="PP_top10")
        
    fig, ax = plt.subplots()

    ax.hlines(temp.University, xmin=0, xmax=temp.PP_top10)
    ax.plot(temp.PP_top10, temp.University, "o", color='black')
    ax.set_xlim(0, .18)
    ax.set_title(rename_field(field))
    
    fp = f"../viz/imgs/top-5-area/{field}-top5.svg".replace(" ", "-")
    plt.savefig(fp)

```

#### Gráfico 2 – Métricas distintas

```python
# Precisamos de um gráfico para cada métrica
columns = [ "PA_F_MF", "PP_OA", "PP_int_collab" ]
for column in columns:
    
    
    
    # Filtra o dataframe
    temp = df [ df.Field == "All sciences"]
    temp = temp.sort_values(by = column, ascending = False)
    temp = temp.head(5).sort_values(by = column)
        
    fig, ax = plt.subplots()

    ax.hlines(temp.University, xmin=0, xmax=temp[column])
    ax.plot(temp[column], temp.University, "o", color='black')
    ax.set_xlim(0, .60)
    ax.set_title(rename_col(column))
    
    fp = f"../viz/imgs/top-5-metric/{rename_col(column)}-top5.svg".replace(" ", "-")
    plt.savefig(fp)

```
